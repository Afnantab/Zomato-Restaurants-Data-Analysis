```python
import matplotlib.pyplot as plt
```

```python
# Approx cost votes
zomato_1=zomato["votes"].head(10)
plt.bar(zomato["approx_cost"][:10],zomato_1)
plt.show()

```

```python
# Top 10 restaurant chains by count
top_chains = zomato['name'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_chains.values, y=top_chains.index, palette="viridis")
plt.title("Top 10 Restaurant Chains in Zomato Dataset")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Name")
plt.show()
```

```python
city_avg_rating = zomato.groupby("location")["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=city_avg_rating.values, y=city_avg_rating.index, palette="magma")
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("Average Rating")
plt.ylabel("location")
plt.show()
```

```python
# Split cuisines and count most common
from collections import Counter

cuisines = zomato['cuisines'].dropna().str.split(",").sum()
cuisine_counts = pd.Series(cuisines).str.strip().value_counts().head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index, palette="coolwarm")
plt.title("Top 15 Popular Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.show()
```

```python
rating_by_online = zomato.groupby("online_order")["rating"].mean().sort_values(ascending=False)
print(rating_by_online)

plt.figure(figsize=(6,4))
sns.barplot(x=rating_by_online.index, y=rating_by_online.values)
plt.title("Average Rating by Online Ordering")
plt.xlabel("Online Order (Yes/No)")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

```python
# Top 10 popular cuisines
plt.figure(figsize=(10,6))
zomato['cuisines'].str.split(',').explode().str.strip().value_counts().head(10).plot(kind='bar', color='coral')
plt.title("Top 10 Popular Cuisines", fontsize=14)
plt.xlabel("Cuisine", fontsize=12)
plt.ylabel("Count of Restaurants", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# avrage rating by resturant type
plt.figure(figsize=(10,6))
sns.barplot(x='rest_type', y='rating', data=zomato, errorbar=None, palette="viridis")
plt.title("Average Rating by Restaurant Type", fontsize=14)
plt.xlabel("Restaurant Type", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Online order vs NO online order 
plt.figure(figsize=(6,5))
sns.boxplot(x="online_order", y="votes", data=zomato, palette="Set1")
plt.title("Votes Distribution: Online Order vs No Online Order", fontsize=14)
plt.xlabel("Online Ordering Available", fontsize=12)
plt.ylabel("Votes", fontsize=12)
plt.show()
```

```python
# Destribution of resturant rating
plt.figure(figsize=(8,5))
sns.histplot(zomato['rating'], bins=20, kde=True, color="purple")
plt.title("Distribution of Restaurant Ratings", fontsize=14)
plt.xlabel("Ratings", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
# Average rate across price ranges
plt.figure(figsize=(8,5))
sns.barplot(x="online_order", y="rating", data=zomato, palette="coolwarm", errorbar=None)
plt.title("Average Rating Across Price Ranges", fontsize=14)
plt.xlabel("Price Range", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.show()
```

```python
# Top 15 cities with most resturants
plt.figure(figsize=(12,6))
zomato['location'].value_counts().head(15).plot(kind='bar', color="teal")
plt.title("Top 15 Cities with Most Restaurants", fontsize=14)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Table booking availbility ratings
plt.figure(figsize=(6,5))
sns.boxplot(x="book_table", y="rating", data=zomato, palette="Set2")
plt.title("Table Booking Availability vs Ratings", fontsize=14)
plt.xlabel("Table Booking", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
```
